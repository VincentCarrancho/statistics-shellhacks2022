# ShellHacks 2022 Demographic Data

This data set is based on the CSV coming from the ShellHacks admin website.

```python
df = pd.read_csv("dataset1.csv", encoding='utf-8')
df
```

## Filtering the data between the time period of July 2nd to September 14th

```python
df['Time Created'] = pd.to_datetime(df['Time Created'])
datefiltered_df = df.loc[(df['Time Created'] >= '2022-07-01') & (df['Time Created'] < '2022-09-14')]
datefiltered_df
```

## Total In-Person Hispanic Population Calculation

We are filtering the dataset by if they are remote and cutting out the data sets that preferred not to answer.

```python
inperson_df = datefiltered_df[(datefiltered_df['Attendance'] == 'In-Person') & (datefiltered_df['Ethnicity'] != 'Prefer not to answer')]
totalInPerson = inperson_df['Time Created'].count()
totalInPerson
```

```python
totalHispanic_df = inperson_df[inperson_df['Ethnicity'] == 'Hispanic or Latinx']
totalHispanicInPerson = totalHispanic_df['Time Created'].count()
```

```python
totalHispanicInPerson / totalInPerson
```

## Total Hispanic Population Calculation (Online + In-Person)

```python
totalHackers_demographic_df = datefiltered_df[datefiltered_df["Ethnicity"] != "Prefer not to answer"]
```

```python
totalHackers = totalHackers_demographic_df['Time Created'].count()
```

```python
totalHispanicHackers_df = totalHackers_demographic_df[totalHackers_demographic_df['Ethnicity'] == 'Hispanic or Latinx']
totalHispanicHackers = totalHispanicHackers_df['Time Created'].count()
```

```python
totalHispanicHackers / totalHackers
```
